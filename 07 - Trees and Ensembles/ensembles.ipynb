```python
import numpy as np
input_df[["BloodPressure", "SkinThickness", "Insulin", "BMI", "Glucose"]] = input_df[["BloodPressure", "SkinThickness", "Insulin", "BMI", "Glucose"]].replace(0, np.nan)
```

```python
input_df.describe()
```

```python
input_df["BloodPressure"].fillna(input_df["BloodPressure"].mean(), inplace=True)
input_df["SkinThickness"].fillna(input_df["SkinThickness"].mean(), inplace=True)
input_df["Insulin"].fillna(input_df["Insulin"].mean(), inplace=True)
input_df["BMI"].fillna(input_df["BMI"].mean(), inplace=True)
input_df["Glucose"].fillna(input_df["Glucose"].mean(), inplace=True)
```

```python
x = input_df.drop("Outcome", axis=1)
y = input_df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=3, random_state=0)
model = model.fit(x_train, y_train)
```

```python
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 20))
plot_tree(model, filled=True, feature_names=x.columns, class_names=["diabetes", "healthy"])
plt.show()
```

```python

model = DecisionTreeClassifier(max_depth=6, random_state=0, ccp_alpha=0.01)
model = model.fit(x_train, y_train)
```

```python
plt.figure(figsize=(20, 20))
plot_tree(model, filled=True, feature_names=x.columns, class_names=["diabetes", "healthy"])
plt.show()
```

```python
# typo forest


from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model = model.fit(x_train, y_train)
```

```python

from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = { "ccp_alpha": [0.001, 0.01, 0.1, 1, 10, 100],
               "criterion": ["gini", "entropy"],
               "max_depth": [3, 5, 7, 9, 11],
               "max_leaf_nodes": [3, 5, 7, 9, 11],
               }

clf = GridSearchCV(RandomForestClassifier(), param_grid)
best_model = clf.fit(x_train, y_train)
```

```python
predictions = best_model.predict(x_test)
cm = confusion_matrix(y_test, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model.classes_)
disp.plot()
```
