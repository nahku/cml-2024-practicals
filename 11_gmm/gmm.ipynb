```python
number_of_samples = 2000
sample_mean = -5
sample_sd = 0.7
```

```python
import numpy as np

sample_data = np.random.normal(sample_mean, sample_sd, number_of_samples)
```

```python
pd.DataFrame(sample_data).plot.kde()
```

```python
import seaborn as sns
sns.kdeplot(data=sample_data)
```

```python
sample_data2 = np.random.normal(1, 0.3, number_of_samples)
```

```python
sns.kdeplot(data=[sample_data, sample_data2])
```

```python
sample_data = np.random.normal(45, 17, 480)
sample_data2 = np.random.normal(110, 18, 600)
combined_samples = np.concatenate([sample_data, sample_data2])
sns.kdeplot(data=[combined_samples, mixed_population_df["The_Data"]])
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, covariance_type='full', tol=0.001)
gmm = gmm.fit(combined_samples.reshape(-1, 1))
```

```python
gmm.means_
```
